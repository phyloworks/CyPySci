### Plotting all the cities I have lived in

```python
m = Basemap(width=12000000,height=9000000,projection='lcc',
            resolution=None,lat_1=45.,lat_2=55,lat_0=50,lon_0=-107.)
# draw a land-sea mask for a map background.
# lakes=True means plot inland lakes with ocean color.
m.drawlsmask(land_color='green',ocean_color='aqua',lakes=True)
lons = [-93.620369,-110.4531551,-96.6712,-97.7431,-77.2016667,-71.038887,-122.258423,-108.7419444,-106.621216,-105.944183,-95.235]
lats = [42.034534,33.3453349,34.8043,30.2672,39.1433333,42.364506,37.8716667,35.5280556,35.113281,35.691544,38.9716667]
x, y = m(lons,lats)
m.plot(x, y, 'wo',)
```

```python
m2 = Basemap(width=12000000,height=9000000,projection='eqdc',
            resolution=None,lat_1=45.,lat_2=55,lat_0=50,lon_0=-107.)
m2.drawlsmask(land_color='coral',ocean_color='aqua',lakes=True)
x, y = m2(lons,lats)
m2.plot(x, y, 'bo',)
```
